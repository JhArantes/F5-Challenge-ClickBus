## Desanonimizando dados

```python
brazilian_cities = pd.read_csv(r'brazilian_cities.csv')

brazilian_cities.head()
```

```python
# Pegar lista de cidades ordenadas por população (mais populosas primeiro)
lista_cidades = brazilian_cities.sort_values(by="osm_population", ascending=False)["city"].tolist()

# Criar um dicionário que mapeia cada hash único para uma cidade
hashes_unicos = csv_clickbus["Destino_da_viagem_de_ida"].unique()
map_hash_para_cidade = {}

for i, h in enumerate(hashes_unicos):
    # Vai pegando da lista de cidades em ordem, e se acabar repete
    map_hash_para_cidade[h] = lista_cidades[i % len(lista_cidades)]

# Substituir no dataframe
csv_clickbus["Destino_da_viagem_de_ida"] = csv_clickbus["Destino_da_viagem_de_ida"].map(map_hash_para_cidade)

csv_clickbus
```
