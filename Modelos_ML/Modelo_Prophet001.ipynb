## Libs

```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv('../desafio clickbus/desafio clickbus/dados_desafio_fiap/pt_clickbus_Trat.csv')
df.head()
```

```python
# Garantir que a coluna está no formato datetime
df['Data_compra'] = pd.to_datetime(df['Data_compra'])

# Pegar o último ano disponível
ultimo_ano = df['Data_compra'].dt.year.max()

# Filtrar apenas esse ano
df_ultimo_ano = df[df['Data_compra'].dt.year == ultimo_ano]

print(f"Último ano encontrado: {ultimo_ano}")
print(df_ultimo_ano.head())
```

- Mudando nome das colunas para o Prophet

```python
# Deixando 1 linha por dia, somando o valor total das compras
df_daily = df.groupby("Data_compra").Valor_total_da_compra.sum().reset_index()

# Renomeando colunas para o padrão exigido pelo Prophet
df_prophet = df[['Data_compra', 'Valor_total_da_compra']].rename(
    columns={'Data_compra': 'ds', 'Valor_total_da_compra': 'y'}
)

df_prophet.head()
```

```python
m = Prophet()
m.fit(df_prophet)
```

```python
future - m.make_future_dataframe(periods=30)
forecast = m.predict(future)
```
